## The Format of this Lesson

This lesson draws on a [Software Carpentry lesson on Programming with Python](http://swcarpentry.github.io/python-novice-gapminder/) for its structure. [The Carpentries](https://carpentries.org) is a diverse global community teaching foundational coding and data science skills to researchers.

The Python lesson is typically taught over a full day using "live coding", where learners follow along with the Instructor, entering and executing code, and discussing issues as they arise. We don't have time for that approach today, but feel free to try things out as we go along.

Many Universities and other research organisations, including St Andrews, run regular 2-day Software Carpentry workshops.

Hopefully, today's lesson will equip you to understand and adapt code you find online e.g. in the [sunpy examples](https://docs.sunpy.org/en/stable/generated/gallery/index.html) as well as to write code of your own.
